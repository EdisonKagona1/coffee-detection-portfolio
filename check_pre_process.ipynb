```python
import os
from os import path

import cv2
import matplotlib.pyplot as plt
from matplotlib import patches
from matplotlib.patches import Rectangle
%matplotlib inline 

from utils import display_image_and_box

import time
```

```python
images_dir = "coffee_raw/Imagenes/"

for i, image in enumerate(os.listdir(images_dir)):

    if image[-3:] == "txt":
        continue

    path_image = path.join(images_dir, image)
    img_cv = cv2.imread(path_image)
    img_cv = cv2.resize(img_cv, dsize=(500,500))
    #print(img)

    if i > 11:
        print(path_image)
        plt.imshow(img_cv)
        break
```

```python
######## DISPLAY RANDOM IMAGES ###########
#######################################
import random

images_dir = "coffee_raw/Imagenes/"

all_images = os.listdir(images_dir)
random.shuffle(all_images) # inplace

for i, image_path in enumerate(all_images):

    if image_path[-3:] == "txt":
        continue

    image_path_full = path.join(images_dir, image_path)
    img = cv2.imread(image_path_full)
    
    associated_txt = image_path[:-3] + "txt"
    associated_txt_full_path = path.join(images_dir, associated_txt)

    bbox_file = open(associated_txt_full_path, "r")

    print(image_path)
    display_image_and_box(img, bbox_file, name=image_path)
    time.sleep(1)

    if i > 10:
        break
```

```python
######## DISPLAY SPECIFIC IMAGES ######
#######################################

images_dir = "coffee_raw/Imagenes/"

images_to_dislay = {"1615402491282c1.jpg"}

for i, image_path in enumerate(images_to_dislay):

    if image_path[-3:] == "txt":
        continue

    if image_path not in images_to_dislay:
        continue

    image_path_full = path.join(images_dir, image_path)
    img = cv2.imread(image_path_full)
    
    associated_txt = image_path[:-3] + "txt"
    associated_txt_full_path = path.join(images_dir, associated_txt)

    bbox_file = open(associated_txt_full_path, "r")

    display_image_and_box(img, bbox_file, name=image_path, save=True)
```
