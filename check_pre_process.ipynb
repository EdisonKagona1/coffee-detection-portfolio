```python
def display_image_and_box(image, bbox_file, name):

    height, width = img.shape[0], img.shape[1]

    # Display the image

    fig, (ax1, ax2) = plt.subplots(1, 2)

    ax1.imshow(img)
    ax2.imshow(img)

    ax2 = plt.gca()

    for bbox_str in bbox_file.readlines():
        object_class, center_x, center_y, width_x, width_y = tuple(map(float, bbox_str.split(" ")))
        anchor_x = (center_x - width_x / 2) * width
        anchor_y = (center_y - width_y / 2) * height
        width_x *= width
        width_y *= height

        ax2.add_patch(
            patches.Rectangle(
                (anchor_x, anchor_y),
                width_x,
                width_y,
                fill=True
            ) ) 
    
    plt.title(name)
    plt.show()
    time.sleep(1)
```
